```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
```

```python
train_data = pd.read_csv('train_data.txt', sep=':::', header=None, names=['ID', 'Title', 'Genre', 'Description'])
test_data = pd.read_csv('test_data_solution.txt', sep=':::', header=None, names=['ID', 'Title', 'Genre', 'Description'])
```

```python
X_train = train_data[['Title', 'Description']]
y_train = train_data['Genre']
X_test = test_data[['Title', 'Description']]
y_test = test_data['Genre']
```

```python
vectorizer = TfidfVectorizer()
X_train_vectorized = vectorizer.fit_transform(X_train['Title'] + ' ' + X_train['Description'])
X_test_vectorized = vectorizer.transform(X_test['Title'] + ' ' + X_test['Description'])
```

```python
svm_model = SVC(kernel='linear', C=1.0)
svm_model.fit(X_train_vectorized, y_train)
```

```python
y_pred = svm_model.predict(X_test_vectorized)
y_pred
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```
